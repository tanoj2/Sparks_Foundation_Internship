### Handling Outliers

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25, 0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
lowp, upperp = remove_outlier(df['Sales'])
df['Sales'] = np.where(df['Sales']>upperp, upperp, df['Sales'])
df['Sales'] = np.where(df['Sales']<lowp, lowp, df['Sales'])
```

### Boxplot for Sales after removing outliers

```python
df.boxplot(column = "Sales")
plt.show()
```

### Handling Outliers

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25, 0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
lowp, upperp = remove_outlier(df['Profit'])
df['Profit'] = np.where(df['Profit']>upperp, upperp, df['Profit'])
df['Profit'] = np.where(df['Profit']<lowp, lowp, df['Profit'])
```
